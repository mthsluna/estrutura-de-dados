```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

G.add_nodes_from([
    "David Bowie", "Prince", "Michael Jackson", "Madonna",
    "Lady Gaga", "Beyonce", "Kanye West", "Rihanna", "Jay-Z"
])

G.add_edges_from([
    ("David Bowie", "Prince"),
    ("Prince", "Michael Jackson"),
    ("David Bowie", "Madonna"),
    ("Madonna", "Lady Gaga"),
    ("Michael Jackson", "Beyonce"),
    ("Beyonce", "Rihanna"),
    ("Jay-Z", "Rihanna"),
    ("Jay-Z", "Kanye West"),
    ("Kanye West", "Rihanna")
])

print(f"Nós do Grafo : {G.nodes()}")
print(f"Arestas do Grafo : {G.edges()}")

plt.figure(figsize=(10, 7))

pos = nx.spring_layout(G, k=1.2, seed=52)

nx.draw(G, pos, with_labels=True, node_color="orange",
        node_size=3000, font_color="black", font_size=8,
        font_weight='bold', arrowsize=15, width=1.5)

plt.title("Grafo de Influência entre Pessoas Famosas")
plt.show()
```
